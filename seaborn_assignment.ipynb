Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ans-Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots you can create using Seaborn, along with their typical use cases:

1.Scatter Plot:
   - Use: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying      patterns, clusters, or trends in the data.
   - Seaborn Function: `sns.scatterplot()`

2.Line Plot:
   - Use: Line plots are used to display the trends and changes in data over a continuous interval, such as time. They are            suitable for visualizing data with ordered categories.
   - Seaborn Function: `sns.lineplot()`

3.Bar Plot:
   - Use: Bar plots are used to display and compare the values of categorical data. They are useful for visualizing counts,            frequencies, or proportions of different categories.
   - Seaborn Function: `sns.barplot()`

4.Histogram:
   - Use: Histograms are used to visualize the distribution of a single continuous variable. They help in understanding the            data's central tendency, spread, and skewness.
   - Seaborn Function: `sns.histplot()`

5.Box Plot (Box-and-Whisker Plot):
   - Use: Box plots are used to display the summary statistics (median, quartiles) and identify outliers in a dataset. They are        particularly useful for comparing the distribution of data across different categories.
   - Seaborn Function: `sns.boxplot()`

These are just a few of the many types of plots that Seaborn can create. Seaborn also provides support for more advanced plots like violin plots, pair plots, heatmaps, and more, making it a powerful tool for data visualization in Python. The choice of plot depends on the nature of your data and the insights you want to gain from it.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal)
```

```python
fmri['event'].value_counts()
```

```python
fmri['region'].value_counts()
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, data = fmri,hue='event')
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, data = fmri,style='region') 
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, data = fmri,style='region', hue='event') 
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
import matplotlib.pyplot as plt
```

```python


# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.heatmap(flights_pivot, annot=True, cmap="YlGnBu", fmt="d")

# Add labels and a title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap (Flights Dataset)")

```
